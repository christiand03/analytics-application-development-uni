```python
import pandas as pd
```

```python
df2 = pd.read_parquet("C:/Users/paulh/Desktop/projects/AAD/analytics-application-development-uni/Positionsdaten.parquet")
```

```python
df2.info(memory_usage="deep")
```

```python
# Schritt 1: Finde die Namen aller Spalten vom Typ 'object'
object_spalten = df2.select_dtypes(include=['object']).columns

print(f"Gefundene Object-Spalten: {list(object_spalten)}\n")

# Schritt 2: Iteriere (loope) durch diese Spalten und gib die Anzahl der Uniques aus
for spalte in object_spalten:
    # .nunique() z√§hlt die Anzahl der einzigartigen Werte in einer Spalte (Series)
    anzahl = df2[spalte].nunique()
    print(f"Spalte '{spalte}': {anzahl} einzigartige Werte")
```

```python
# Deine definierte Liste von Spalten
spalten_zum_umwandeln = ['KvaRechnung_ID', 'KvaRechnung_Nummer', 'Mengeneinheit', 'Bemerkung']

# Wende .astype('category') nur auf die Spalten in dieser Liste an
df2[spalten_zum_umwandeln] = df2[spalten_zum_umwandeln].astype('category')

object_columns = df2.select_dtypes('object').columns
df2[object_columns] = df2[object_columns].astype('string')

df2.info(memory_usage='deep')
```

```python
# Downcast integer columns
int_cols = df2.select_dtypes(include=['int64']).columns
for col in int_cols:
    df2[col] = pd.to_numeric(df2[col], downcast='integer')

# Downcast float columns
float_cols = df2.select_dtypes(include=['float64']).columns
for col in float_cols:
    df2[col] = pd.to_numeric(df2[col], downcast='float')

print("\n--- Memory Usage After Numeric Downcasting ---")
df2.info(memory_usage='deep')
```

```python
df2.replace('-', None, inplace=True)
df2.info(memory_usage='deep')
```
