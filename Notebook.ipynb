```python
df_unique = df[["DH_ID", "Land", "PLZ_SO", "PLZ_HW", "PLZ_VN", "Schadenart_Name", "Falltyp_Name", "Gewerk_Name", "Kundengruppe", "Handwerker_Name"]]
```

```python
for col in df_unique.columns:
    
    print(f"--- Tabelle für Spalte: '{col}' ---")
    
    # 1. Eindeutige Werte aus der Spalte extrahieren und leere Werte (NaN) entfernen
    unique_values = df_unique[col].unique()
    
    # 2. Aus diesen Werten einen neuen, kleinen DataFrame erstellen
    # Die Spalte in diesem neuen DataFrame bekommt den Namen der Originalspalte
    tabelle_pro_spalte = pd.DataFrame(unique_values, columns=[col])
    
    # 3. Die neu erstellte Tabelle ausgeben
    if len(tabelle_pro_spalte) > 20:
        display(tabelle_pro_spalte)
    else:
        print(tabelle_pro_spalte)
    
    # 4. Eine Leerzeile für bessere Lesbarkeit zwischen den Tabellen
    print("\n")
```

```python
from IPython.display import display, HTML

# Hier kannst du die maximale Höhe der Tabellen in Pixeln festlegen
MAX_TABLE_HEIGHT_PX = 300

for col in df_unique.columns:
    
    # Eine schöne Überschrift für jede Tabelle
    display(HTML(f"<h3>Tabelle für Spalte: '{col}'</h3>"))
    
    # Eindeutige Werte holen und NaN entfernen
    unique_values = df_unique[col].unique()
    
    # Neuen DataFrame erstellen
    tabelle_pro_spalte = pd.DataFrame(unique_values, columns=[col])
    
    # --- DER MAGISCHE TEIL ---
    # 1. Wandle den DataFrame in einen HTML-String um.
    #    index=False verhindert, dass die Index-Spalte (0, 1, 2...) im HTML erscheint.
    html_table = tabelle_pro_spalte.to_html(index=False)
    
    # 2. Erstelle einen scrollbaren Container um die HTML-Tabelle.
    #    - max-height: Die Tabelle wird nicht höher als dieser Wert.
    #    - overflow-y: auto: Wenn der Inhalt höher ist, zeige eine vertikale Scrollleiste an.
    scrollable_html = f"""
    <div style="max-height: {MAX_TABLE_HEIGHT_PX}px; overflow-y: auto; border: 1px solid #e0e0e0; padding: 5px;">
        {html_table}
    </div>
    """
    
    # 3. Zeige das fertige HTML-Konstrukt an.
    display(HTML(scrollable_html))
    
    # Optional: Eine Trennlinie für die Übersichtlichkeit
    display(HTML("<hr>"))
```
