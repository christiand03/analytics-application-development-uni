```python
import pandas as pd
```

```python
df = pd.read_parquet("C:/Users/paulh/Desktop/projects/AAD/analytics-application-development-uni/Auftragsdaten.parquet")
```

```python
df.info(memory_usage='deep')
```

```python
# Schritt 1: Finde die Namen aller Spalten vom Typ 'object'
object_spalten = df.select_dtypes(include=['object']).columns

print(f"Gefundene Object-Spalten: {list(object_spalten)}\n")

# Schritt 2: Iteriere (loope) durch diese Spalten und gib die Anzahl der Uniques aus
for spalte in object_spalten:
    # .nunique() z√§hlt die Anzahl der einzigartigen Werte in einer Spalte (Series)
    anzahl = df[spalte].nunique()
    print(f"Spalte '{spalte}': {anzahl} einzigartige Werte")
```

```python
# Deine definierte Liste von Spalten
spalten_zum_umwandeln = ['Land', 'PLZ_SO', 'PLZ_HW', 'PLZ_VN', 'address1_postalcode', 
                       'Schadenart_Name', 'Falltyp_Name', 'Gewerk_Name', 'Kundengruppe', 
                       'Handwerker_Name']

# Wende .astype('category') nur auf die Spalten in dieser Liste an
df[spalten_zum_umwandeln] = df[spalten_zum_umwandeln].astype('category')

object_columns = df.select_dtypes('object').columns
df[object_columns] = df[object_columns].astype('string')
```

```python
df.info(memory_usage='deep')
```

```python
# Downcast integer columns
int_cols = df.select_dtypes(include=['int64']).columns
for col in int_cols:
    df[col] = pd.to_numeric(df[col], downcast='integer')

# Downcast float columns
float_cols = df.select_dtypes(include=['float64']).columns
for col in float_cols:
    df[col] = pd.to_numeric(df[col], downcast='float')

print("\n--- Memory Usage After Numeric Downcasting ---")
df.info(memory_usage='deep')# Downcast integer columns
int_cols = df.select_dtypes(include=['int64']).columns
for col in int_cols:
    df[col] = pd.to_numeric(df[col], downcast='integer')

# Downcast float columns
float_cols = df.select_dtypes(include=['float64']).columns
for col in float_cols:
    df[col] = pd.to_numeric(df[col], downcast='float')

print("\n--- Memory Usage After Numeric Downcasting ---")
df.info(memory_usage='deep')
```
